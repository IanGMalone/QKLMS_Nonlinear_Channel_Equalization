## Implementation

##### 1. Simulate the dynamics provided by the above equations and generate a data-set of 5000 points. The input signal is Gaussian random variable with zero mean and unit variance.

```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import signal
```

```python
# input signal:
# 5000 points of Gaussian random variable with zero mean and unit variance
number_samples = 5000
mu = 0
std = 1
yn = np.random.normal(mu,std,number_samples)

# linear filter
filt = [-0.8, 0.7]
tn = np.convolve(filt, yn)

#memoryless nonlinearity
qn = tn + 0.25*tn**2 + 0.11*tn**3

# 15dB AWGN added to signal
db = 15
power = 10**(db/10)
xn = qn + np.random.normal(0,1*power,len(qn))

# desired response is yn delayed by some value D
D = 2
dn = np.pad(yn, (D,0))[:-2]
```

```python
plt.figure(figsize=(10,5))
plt.plot(xn, label='xn')
plt.plot(qn, label='qn')
plt.legend(loc="upper right")
plt.xlabel("Samples")
```
